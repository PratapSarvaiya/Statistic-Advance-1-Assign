# Given a sample mean of 105, a population mean of 100,a standard deviation of 15, and a sample size of 25,calculate the z-score and p-value.Based on a significance level of 0.05,do you reject or fail to reject the null hypothesis? Task:Write python code to calculate the z-score and p-value for the given data. Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.

```python
import scipy.stats as stats
import math
sample_mean = 105
population_mean = 100
std_dev = 15
sample_size = 25
alpha = 0.05

# calculate the z-score
z_score = (sample_mean- population_mean) / (std_dev / math.sqrt(sample_size))

# calculate the p-value(two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Decision based on the significance level
if p_value < alpha:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

print(f"Z-score: {z_score}")
print(f"p-value: {p_value}")
print(f"Decision: {decision}")
```

# Question 10:

# Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using python.Generate 1000 samples and plot the distribution.What is the expected mean and variance? Task: Use python to generate the data,plot the distribution,and calculate the mean and variance. objective:Understand the properties of a binomial distribution and verify them through simulation.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

n_trials = 10
p_success = 0.6
n_samples = 1000

# Generate binomial distribution samples
binomial_samples = np.random.binomial(n=n_trials, p=p_success, size=n_samples)

# plot the distribution
sns.histplot(binomial_samples, bins=range(n_trials + 2), kde=False, color='blue', stat='density')
plt.title('Binomial Distribution (n=10, p=0.6)')
plt.xlabel('Number of Successes')
plt.ylabel('Density')
plt.show()

# Calculate expected mean and Variance
expected_mean = n_trials * p_success
expected_variance = n_trials * p_success * (1 - p_success)

# Calculate sample mean and variance
sample_mean = np.mean(binomial_samples)
sample_variance = np.var(binomial_samples)

expected_mean, expected_variance, sample_mean, sample_variance
```
