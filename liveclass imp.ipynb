```python
import numpy as np
import matplotlib.pyplot as plt

# Generate Bernoulli distribution
s = np.random.Bernoulli(n=1, p=0.5, size=1000)

# Plot histogram
plt.hist(s, density=True)
plt.title("The Bernoulli Distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate Bernoulli distribution
s = np.random.binomial(n=1, p=0.5, size=1000)

# Plot histogram
plt.hist(s, density=True)
plt.title("The Bernoulli Distribution")
plt.show()
```

```python
# poisoin distribution
sample = np.random.poisson(lam = 5, size = 10000)
plt.hist(sample, bins = 20)
plt.title("Poisson distribution")
plt.show()
```

```python
# continuous distributions >> normal/gaussian distribution
np.random.normal(0, 1, size = 1000)
sns.distplot(sample, bins = 30, kde = True)
plt.show()
```

```python
# uniform distribution
sample = np.random.uniform(-1,0,1000)
plt.hist(sample, bins=20,desity=True)
plt.show()
```

```python
# chisquare distribution
sample = np.random.chisquare(df = 1, size=1000)
plt.hist(sample, bins=100, desity=True)
plt.show()
```

```python
sample = np.random.chisquare(df = 1, size=100)
sample = sample[np.where(abs(sample)<5)]
plt.hist(sample, bins=100, desity=True)
plt.show()
```

```python
# f distribution
sample = np.random.f(dfnum = 1,dfden=1, size=1000)
plt.hist(sample, bins=100, desity=True)
plt.show()
```
